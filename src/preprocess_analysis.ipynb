```python
####
#
# Randomly copy 1000 files
#
####


import random
import shutil
import os

rootdir = '../data/pert'
outdir = '../data/pert_1000/'

ref = 1000

dirsAndFiles = {}   # here we store a structure  {folder: [file1, file2], folder2: [file2, file4] }
dirs = [x[0] for x in os.walk(rootdir)] # here we store all sub-dirs

for dir in dirs:
    dirsAndFiles[dir] = [f for f in os.listdir(dir) if os.path.isfile(os.path.join(dir, f))]
    
for (dir, files) in dirsAndFiles.iteritems():
    for i in xrange(ref):
        fe = random.choice(files)
        files.remove(fe)
        shutil.copy(os.path.join(dir, fe), outdir)
```

```python
####
#
# csv
#
####

import pandas as pd
from os import listdir
from os.path import isfile, join
import numpy as np

#outdir = '../data/pert_EGFR/'
outdir = '../data/pert_1000/'
pert_files = [f for f in listdir(outdir) if isfile(join(outdir, f))]

experiments = []
num_of_experiments = []

experiments_cell_lines = set()
experiments_pert_plate = set()
experiments_hour = set()
experiments_dose = set()

for pert_file in pert_files:
    pert_frame = pd.read_csv(os.path.join(outdir, pert_file))
    num_of_experiments.append(len(pert_frame.columns)-1)

    for pert in pert_frame.columns:
        pert_split = pert.split(":")
        
        if len(pert_split) != 3:
            continue
        
        #an environment of a perterbation
        pert_env = pert_split[0]
        
        pert_env_split = pert_env.split("_")        
        
        ## a plate used for a perturbation
        pert_plate = pert_env_split[0]
        
        ## a plate used for a perturbation
        pert_cell_line = pert_env_split[1]
        
        ## a plate used for a perturbation
        pert_hour = pert_env_split[2]
        
        #a drug name used for a perturabation
        drug_name = pert_split[1]
        
        #a dose of a perterbuation
        pert_dose = pert_split[2]
        
        experiments.append(pert)
        experiments_cell_lines.add(pert_cell_line)
        experiments_pert_plate.add(pert_plate)
        experiments_hour.add(pert_hour)
        experiments_dose.add(pert_dose) 

print "\n\n>>> Experiment Level Analysis"
print "Mean of Experiments\t", np.mean(num_of_experiments)
print "Deviation of Experiments\t", np.std(num_of_experiments)
print "Maximum Number of Experiments\t", np.max(num_of_experiments)
print "Minimum Number of Experiments\t", np.min(num_of_experiments)
        
print "\n\n>>> Cell Line Level Analysis"
print "Unique Cell Lines\t", len(experiments_cell_lines)
print "Cell Lines\t", experiments_cell_lines

print "\n\n>>> Plate Level Analysis"
print "Unique Plate\t", len(experiments_pert_plate)
print "Plate\t", experiments_pert_plate

print "\n\n>>> Hour Level Analysis"
print "Unique Hour\t", len(experiments_hour)
print "Hour\t", experiments_hour

print "\n\n>>> Dose Level Analysis"
print "Unique Dose\t", len(experiments_dose)
print "Dose\t", experiments_dose
```
